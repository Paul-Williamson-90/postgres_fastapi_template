```python
import requests
from uuid import uuid4

new_id = str(uuid4())
new_post_response = requests.post(
    url="http://localhost:8000/posts/create",
    json={
        "id": new_id,
        "title": "New Post",
        "content": "This is a new post.",
        "published": True,
        "created_at": "2021-01-01T00:00:00",
    }
)
new_post = new_post_response.json()
new_id = new_post["id"]

response = requests.get(
    url=f"http://localhost:8000/posts/{new_id}"
)
response.status_code
```
